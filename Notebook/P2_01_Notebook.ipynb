```python
import math

### Remplissage des colonnes vides

for country in countryList:
    for i in reversed(range(0,len(dataTri_df.columns))) :
        if isinstance(secondary_df.iloc[countryList.index(country),i], str):
            break
        elif math.isnan(secondary_df.iloc[countryList.index(country),i]) == True :
            continue
        else:
            analyse_df.iloc[countryList.index(country),3] = secondary_df.iloc[countryList.index(country),i]
            break
    for i in reversed(range(0,len(dataTri_df.columns))) :
        if isinstance(tertiary_df.iloc[countryList.index(country),i], str):
            break
        if math.isnan(tertiary_df.iloc[countryList.index(country),i]) == True :
            continue
        else :
            analyse_df.iloc[countryList.index(country),4] = tertiary_df.iloc[countryList.index(country),i]
            break
    if math.isnan(analyse_df.iloc[countryList.index(country),3]) == True:
        analyse_df.iloc[countryList.index(country),5] = analyse_df.iloc[countryList.index(country),4]            
    elif math.isnan(analyse_df.iloc[countryList.index(country),4]) == True:
        analyse_df.iloc[countryList.index(country),5] = analyse_df.iloc[countryList.index(country),3]        
    else:
        analyse_df.iloc[countryList.index(country),5] = analyse_df.iloc[countryList.index(country),3] + analyse_df.iloc[countryList.index(country),4]
    for i in reversed(range(0,len(dataTri_df.columns))) :
        if isinstance(internet_df.iloc[countryList.index(country),i], str):
            break
        if math.isnan(internet_df.iloc[countryList.index(country),i]) == True :
            continue
        else :
            analyse_df.iloc[countryList.index(country),6] = internet_df.iloc[countryList.index(country),i]
            break


analyse_df.to_csv('analyse.csv', index = False)
```
